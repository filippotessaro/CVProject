```python
import pandas as pd
df = pd.read_csv('giacomo-measuresdataframe.csv')
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.kdeplot(df['height']).set_title("Distribution of Height in mm")
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['dubois'])
sns.kdeplot(df['mosteller']).set_title("Weight Distribution in KG")
```

```python
plt.savefig('books_read.png')
```

```python
ax = sns.lineplot(data=df)
```

```python
#plt.plot(x, x, label='linear')
import matplotlib.pyplot as plt

plt.plot(df['dubois'].index, df['dubois'], label='DuBois')
plt.plot(df['dubois'].index, df['mosteller'], label='Mean')

#plt.plot(x, df['mosteller'], label='cubic')

plt.xlabel('# frame')
plt.ylabel('Weight KG')

plt.title("DuBois")

plt.legend()

plt.show()
```
